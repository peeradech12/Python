```python
def bfs(visited, graph, node,final): 
  visited.append(node)
  queue.append(node)
  while queue: 
    s = queue.pop(0)
    print (s, end = ' ')
    if s == final :
      break 
    for neighbour in graph[s]:
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)

# Driver Code
graph = {
'0' : ['1','3'],
'1' : ['0','2'],
'2' : ['1','3'],
'3' : ['0','2'],
'4' : ['5','6'],
'5' : ['4','6'],
'6' : ['4','5'],
'7' : ['8'],
'8' : ['7'],
}
visited = [] # List to keep track of visited nodes.
queue = [] #Initialize a queue
value = input('Input initial Node : ')
final= input('Input final Node : ')
print("Initial node : ",value)
print("Final node : ",final)
bfs(visited, graph, value,final)
```
