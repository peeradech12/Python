```python
from urllib.request import urlopen as req
from bs4 import BeautifulSoup as Soup
```

```python
import pandas as pd
import bs4
import requests
import openpyxl
```

```python
name_list = []
Description_list = []

url = 'https://astronomy.swin.edu.au/cosmos/c/Constellation'

webopen = req(url)
page_html = webopen.read()
webopen.close()
data = Soup(page_html,'html.parser')
temp = data.findAll('td')
n = 28
while n <556 :
  result1 = temp[n].text
  result2 = temp[n+2].text
  name_list.append(result1)
  Description_list.append(result2)
  n+= 6

table = pd.DataFrame([name_list,Description_list]).transpose()
table.columns = ['name','Description']
table.set_index('name')
table.to_excel('Constellation.xlsx',engine='openpyxl')
```
